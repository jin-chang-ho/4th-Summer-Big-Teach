# 빈도분석 연습문제

```python
# 1. 데이터를 pd.read_csv() 함수로 읽으시오.
import numpy as np
import pandas as pd

data = pd.read_csv("C:\\Users\\Windows\\Desktop\\대학교\\4학년 여름방학\\Big_AI\\자연어처리\\Final_Dataset\\뉴스데이터.csv", encoding = "cp949")

print(data[:10])
print(len(data))
print(type(data))
```

```python
# 2. 데이터에서 결측치가 있는 행을 제거하시오.

print(data.isnull().sum())
data = data.dropna(axis = 0)
print(data.isnull().sum())
```

```python
# 3. 데이터에서 [분류 2]는 제거한 복사본 data1을 만드시오.
data1 = data.iloc[:, [0, 1, 2, 4]]
print(data1)
```

```python
# 4. 데이터에서 [분류 1]을 제거한 복사본 data2를 만드시오.
data2 = data.iloc[:, [0, 1, 3, 4]]
print(data2)
```

```python
# 5. data1에서 분류 1을 기준으로 텍스트의 어휘 빈도 특징을 조사하시오.

from collections import Counter

Upper_Series = data1.iloc[:, 2]
Upper_List = list(Upper_Series)

upper = []
lower = []

for d in Upper_List:
  d = d.split(",")
  upper.append(d[0])
  lower.append(d[1])

print(upper[:10])
print(lower[:10])

upper_freq = Counter(upper) # 지역이 가장 많고 문화가 가장 적다.
lower_freq = Counter(lower) # 축구가 가장 많다. 작은 거 중엔 1, 2도 많다.

print(upper_freq)
print(lower_freq)
```

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

sorted_keys_sample = sorted(upper_freq, key = upper_freq.get, reverse = True)
sorted_values_sample = sorted(upper_freq.values(), reverse = True)

font_location = "C:\\Windows\\Fonts\\HMFMPYUN.ttf"
font_name = fm.FontProperties(fname = font_location).get_name()
matplotlib.rc("font", family = font_name)

plt.bar(range(len(upper_freq)), sorted_values_sample)
plt.xticks(range(len(upper_freq)), sorted_keys_sample)
plt.show()
```

```python
import matplotlib.pyplot as plt

sorted_keys_sample = sorted(lower_freq, key = lower_freq.get, reverse = True)
sorted_values_sample = sorted(lower_freq.values(), reverse = True)

plt.bar(range(10), sorted_values_sample[:10])
plt.xticks(range(10), sorted_keys_sample[:10])
plt.show()
```

```python
# 6. data2에서 분류 2를 기준으로 텍스트의 어휘 빈도 특징을 조사하시오.

from collections import Counter

Upper_Series = data2.iloc[:, 2]
Upper_List = list(Upper_Series)

upper = []
lower = []

for d in Upper_List:
  d = d.split(",")
  upper.append(d[0])
  lower.append(d[1])

print(upper[:10])
print(lower[:10])

upper_freq = Counter(upper) # 지역이 가장 많고 정치가 가장 적다.
lower_freq = Counter(lower) # 산업_기업이 가장 많다. 가장 작은 값이 2이다.

print(upper_freq)
print(lower_freq)
```

```python
import matplotlib.pyplot as plt

plt.bar(range(len(upper_freq)), sorted(upper_freq.values(), reverse = True))
plt.xticks(range(len(upper_freq)), sorted(upper_freq, key = upper_freq.get, reverse = True))
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.bar(range(10), sorted(lower_freq.values(), reverse = True)[:10])
plt.xticks(range(10), sorted(lower_freq, key = lower_freq.get, reverse = True)[:10])
plt.show()
```

```python
# 7. 2018년 자료와 2019년 자료의 어휘의 빈도별 특징을 구하시오.

year_2018 = []
year_2019 = []
year_another = []

for i in range(len(data)):
  if str(data.iloc[i].날짜).startswith("2018"):
    year_2018.append(str(data.iloc[i].날짜))
  elif str(data.iloc[i].날짜).startswith("2019"):
    year_2019.append(str(data.iloc[i].날짜))

print(year_2018)
print(year_2019)
print(year_another)
print(len(year_2018))
print(len(year_2019))
```

```python
# 3. data1, data2를 각각 8:2로 훈련데이터와 테스트데이터로 구분하시오.

from sklearn.model_selection import train_test_split
from collections import Counter

data1_upper_data = data1_upper.loc[:, "텍스트"]
data1_upper_label = data1_upper.loc[:, "분류1_상위"]

print(data1_upper_data[:10])
print(data1_upper_label[:10])

data1_train_data, data1_test_data, data1_train_label, data1_test_label = train_test_split(data1_upper_data, data1_upper_label, test_size = 0.2, stratify = data1_upper_label)

data1_train_count = Counter(data1_train_label)
data1_test_count = Counter(data1_test_label)

print(data1_train_count)
print(data1_test_count)
```

```python
# 4. data1, data2의 [분류1/2_상위] 컬럼 내용에 따라 자동 분류할 수 있는 딥러닝 모델을 만드시오. & 5. 강의 시간에 학습한 기법을 이용해 기법마다 모델을 만드시오.
def mor_anal(data):
  mor_anal_data = []

  for d in data:
    temp_split_data = rhinoMorph.onlyMorph_list(rn, d, pos = ["NNG", "NNP", "NP", "VV", "VA", "MAG", "IC", "XR", "MM", "MAJ"], eomi = True)
    mor_anal_data.append(temp_split_data)
  
  return mor_anal_data

def to_one_hot(data):
  results = np.zeros((len(data), max_words))
  for i, data in enumerate(data):
    results[i, data] = 1.

  return results

def label_pre1(data):
  temp_data = []
  for d in data:
    if d == "경제":
      temp_data.append(0)
    elif d == "지역":
      temp_data.append(1)
    elif d == "스포츠":
      temp_data.append(2)
    elif d == "IT_과학":
      temp_data.append(3)
    elif d == "사회":
      temp_data.append(4)
    elif d == "문화":
      temp_data.append(5)
    elif d == "국제":
      temp_data.append(6)

  return temp_data

def label_pre2(data):
  temp_data = []
  for d in data:
    if d == "경제":
      temp_data.append(0)
    elif d == "지역":
      temp_data.append(1)
    elif d == "스포츠":
      temp_data.append(2)
    elif d == "IT_과학":
      temp_data.append(3)
    elif d == "사회":
      temp_data.append(4)
    elif d == "문화":
      temp_data.append(5)
    elif d == "국제":
      temp_data.append(6)
    elif d == "정치":
      temp_data.append(7)

  return temp_data
```

```python
import rhinoMorph
rn = rhinoMorph.startRhino()

data1_train_anal = mor_anal(data1_train_data)
data1_test_anal = mor_anal(data1_test_data)
```

```python
from collections import Counter

check_freq = []
for d in data1_train_anal:
  check_freq.extend(d)

print(check_freq)

count = Counter(check_freq)

print(count)
print(len(list(count.keys())))

sum = 0
for index in list(count.keys()):
  if count[index] >= 5:
    sum = sum + 1
print(sum) # max_words는 4000개
```

```python
import numpy as np

data1_train_len = [len(d) for d in data1_train_anal]
print(np.min(data1_train_len))
print(np.max(data1_train_len))
print(np.mean(data1_train_len))
print(np.median(data1_train_len))
print(np.percentile(data1_train_len, [0, 25, 50, 75, 90, 95, 100])) # maxlen은 30
```

```python
from keras.preprocessing.text import Tokenizer
from keras_preprocessing.sequence import pad_sequences
import numpy as np
import math

max_words = 4000
maxlen = 30
class_number = 7

tokenizer = Tokenizer(num_words = max_words)
tokenizer.fit_on_texts(data1_train_anal)
word_index1 = tokenizer.word_index
```

```python
print("%s의 토큰 " % len(word_index1))
```

```python
data1_train_seq = tokenizer.texts_to_sequences(data1_train_anal)
data1_test_seq = tokenizer.texts_to_sequences(data1_test_anal)

print("data1_train_seq 0 : ", data1_train_seq[0])
```

```python
data1_train_pad = pad_sequences(data1_train_seq, maxlen = maxlen)
data1_test_pad = pad_sequences(data1_test_seq, maxlen = maxlen)

print("data1_train_pad : ", data1_train_pad[0])
```

```python
data1_train_one = to_one_hot(data1_train_pad)
data1_test_one = to_one_hot(data1_test_pad)

print("data1_train_one[:100], ", data1_train_one[:100])
```

```python
from collections import Counter

count = Counter(data1_train_label)
print(list(count.keys()))
```

```python
from tensorflow.keras.utils import to_categorical

data1_train_label_pre = label_pre1(data1_train_label)
data1_train_label_pre = to_categorical(data1_train_label_pre)
data1_test_label_pre = label_pre1(data1_test_label)

print("data1_train_label_pre : ", data1_train_label_pre)
print("data1_train_label_pre 0 : ", len(data1_train_label_pre[0]))
```

```python
data1_train_final, data1_val_final, data1_train_label_final, data1_val_label_final = train_test_split(data1_train_one, data1_train_label_pre, test_size = 0.2, stratify = data1_train_label_pre)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(64, activation = "relu", input_shape = (max_words, )))
model.add(Dense(32, activation = "relu"))
model.add(Dense(class_number, activation = "softmax"))
model.compile(optimizer = "rmsprop", loss = "categorical_crossentropy", metrics = ["acc"])
model.summary()
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint

earlystop_callback = EarlyStopping(monitor = "val_acc", patience = 5)
cp_callback = ModelCheckpoint("final_models\\best_model1.h5", monitor = "val_acc", verbose = 1, save_best_only = True)

epochs = 30
batch_size = 32

history = model.fit(data1_train_final, data1_train_label_final, epochs = epochs, batch_size = batch_size, callbacks = [earlystop_callback, cp_callback], validation_data = (data1_val_final, data1_val_label_final))

history_dict = history.history
```

```python
from keras.models import load_model

loaded_model = load_model("final_models\\best_model1.h5")

predict = loaded_model.predict(data1_test_one)
```

```python
import numpy as np

answer = np.argmax(predict, axis = 1)
accuracy = np.sum(np.array(answer == data1_test_label_pre)) / len(answer) * 100
print(accuracy)
```
