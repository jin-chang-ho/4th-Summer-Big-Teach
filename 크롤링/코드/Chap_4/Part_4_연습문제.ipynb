```python
# 1. https://eungdapso.seoul.go.kr/ 에서 시장에게 바란다 게시물 58개 수집하기

#Step 1. 필요한 모듈을 로딩합니다
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup     
import pandas as pd   
import openpyxl
import math
import time  

#Step 2. 사용자에게 검색 관련 정보들을 입력 받습니다.
print("=" *100)
print(" 이 크롤러는 서울 특별시 응답소 사이트의 시장에게 바란다 게시물 수집용 웹크롤러입니다.")
print("=" *100)
collect_cnt = int(input('1.크롤링 할 건수는 몇건입니까?: '))

#Step 3. 수집된 데이터를 저장할 파일 경로 입력받기 
f_dir = "C:\\Users\\Windows\\Desktop\\대학교\\4학년 여름방학\\Big_AI\\크롤링\\코드\\Chap_4\\출력 파일\\"

#Step 4. 크롬 드라이버 설정 및 웹 페이지 열기
s = Service("C:\\chromedriver.exe")
driver = webdriver.Chrome(service=s)

url = 'https://eungdapso.seoul.go.kr/'
driver.get(url)
driver.maximize_window()
time.sleep(2)

#Step 5. 자동으로 바로가기 클릭하기
driver.find_element(By.XPATH,'//*[@id="gnb"]/div[4]/div[2]/div/a').click()
time.sleep(1)

#Step 6. 크롤링 할 총 페이지 번호 출력하기
import math
collect_page_cnt = math.ceil(collect_cnt / 10) # 올림
print('크롤링할 총 페이지 번호 : ', collect_page_cnt)
print('\n')

#Step 7. 각 항목별로 데이터를 추출하여 리스트에 저장하기
no2 = []
title2 = []     #논문제목 저장
day2 = []      #공개일 저장
contents2 = []  #상담내역 저장

no = 1

start = time.time()
for a in range(1, collect_page_cnt + 1) :
    print("%s 페이지 내용 수집 시작합니다 ============================================" % str(a))
    
    inner_check = 11
    if a == collect_page_cnt:
        inner_check = (collect_cnt % 10) + 1

    for temp in range(2, inner_check + 1):
        print("==== %s번째 게시물의 상세내역을 추출합니다 ====" % str(no))
        driver.find_element(By.XPATH, '//*[@id="content_cont"]/div[2]/div/form/div[4]/table/tbody/tr[' + str(temp) + ']/td[2]/a').click()
        time.sleep(1)
        html_1 = driver.page_source
        soup_1 = BeautifulSoup(html_1, 'html.parser')
        
        table_value = soup_1.find("table", "table_style").find_all("td")
        
        title = str(table_value[0])
        title = title.replace("<td>", "")
        title = title.replace("</td>", "")
        
        day = str(table_value[1])
        day = day.replace("<td>", "")
        day = day.replace("</td>", "")
        
        print("1.제목: ", title)
        title2.append(title)
        
        print("2.공개일: ", day)
        day2.append(day)
        
        content = soup_1.find("div", "table_inner_desc").get_text()
        content = content.replace("Q. 상담내용", "")
        content = content.strip()
        
        print("3.상담내역:", content)
        contents2.append(content)
        
        no2.append(no)
        no = no + 1
        
        driver.back()
        
    a += 1 
    b = str(a)
    
    try :
        driver.find_element(By.LINK_TEXT ,'%s' % b).click() 
    except:
        driver.find_element(By.LINK_TEXT, '>').click()

end = time.time()
print("요청하신 작업이 모두 완료되었습니다")

# Step 8. 수집된 데이터를 xls와 csv 형태로 저장하기
import os
import pandas as pd

print("총 소요시간은 %s 초 입니다." % str(end-start))

df = pd.DataFrame()
df['번호']=no2
df['제목']=pd.Series(title2)
df['공개일']=pd.Series(day2)
df['상담내역']=pd.Series(contents2)

n = time.localtime()
s = "%04d-%02d-%02d-%02d-%02d-%02d" % (n.tm_year, n.tm_mon, n.tm_mday, n.tm_hour, n.tm_min, n.tm_sec)
check_dic = f_dir + s + "-서울시응답소"
os.makedirs(check_dic)

ft_name = check_dic + "\\" + s + "-서울시응답소.txt"
fx_name = check_dic + "\\" + s + "-서울시응답소.xls"
fc_name = check_dic + "\\" + s + "-서울시응답소.csv"

# txt 형태로 저장하기
f = open(ft_name, 'w', encoding="UTF-8")

for i in range(1, no):
    f.write("1.제목:"+ title2[i-1]+ "\n")
    f.write("2.공개일:"+ day2[i-1]+"\n")
    f.write("3.상담내역:"+ contents2[i-1]+ "\n")
    f.write("\n")

f.close()
print("파일 저장 완료: txt 파일명 : ", ft_name)

# csv 형태로 저장하기
df.to_csv(fc_name,index=False, encoding="utf-8-sig")
print("파일 저장 완료: csv 파일명 : ", fc_name)

# xls 형태로 저장하기
df.to_excel(fx_name,index=False, encoding="utf-8" , engine='openpyxl')
print("파일 저장 완료: xls 파일명 : ", fx_name)

print('요청하신 데이터 수집 작업이 정상적으로 완료되었습니다')

# 중복여부 확인법 : 복사해서 ctrl+F로 찾아보기
```

```python
# 2. 대한민국 구석구석 사이트에서 제주도 여행 게시물 123개 수집하기

#Step 1. 필요한 모듈을 로딩합니다
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup     
import pandas as pd   
import openpyxl
import math
import time  

#Step 2. 사용자에게 검색 관련 정보들을 입력 받습니다.
print("=" *100)
print(" 이 크롤러는 제주도 여행 사이트의 게시물 수집용 웹크롤러입니다.")
print("=" *100)
query_txt = input('1.수집할 자료의 키워드는 무엇입니까?: ')
collect_cnt = int(input('2. 몇 건의 정보를 수집할까요?: '))

#Step 3. 수집된 데이터를 저장할 파일 경로 입력받기 
f_dir = "C:\\Users\\Windows\\Desktop\\대학교\\4학년 여름방학\\Big_AI\\크롤링\\코드\\Chap_4\\출력 파일\\"

#Step 4. 크롬 드라이버 설정 및 웹 페이지 열기
s = Service("C:\\chromedriver.exe")
driver = webdriver.Chrome(service=s)

url = 'https://korean.visitkorea.or.kr/'
driver.get(url)
driver.maximize_window()
time.sleep(2)

#Step 5. 자동으로 검색어 입력 후 조회하기
element = driver.find_element(By.ID,'inp_search')
driver.find_element(By.ID,'inp_search').click( )
element.send_keys(query_txt)
element.send_keys("\n")
time.sleep(2)

#Step 6. 크롤링 할 총 페이지 번호 알아내기
import math
collect_page_cnt = math.ceil(collect_cnt / 10) # 올림

#Step 7. 각 항목별로 데이터를 추출하여 리스트에 저장하기
no2 = []
title2 = []     #제목 저장
area2 = []      #지자체 저장
hashtag2 = []   #해시태그 저장
contents2 = []  #본문내용 저장

no = 1
xpath_check = 1

for a in range(1, collect_page_cnt + 1) :
    print("%s 페이지 내용 수집 시작합니다 ============================================" % str(a))  
    
    html_1 = driver.page_source
    soup_1 = BeautifulSoup(html_1, 'html.parser')
    
    div_list = soup_1.find_all("div", "area_txt")
    
    for div in div_list:
        print("==== %s번째 정보를 추출하고 있습니다 ====" % str(no))
        
        print("1.번호 : ", no)
        no2.append(no)
        
        title = div.find('div', 'tit').find("a").get_text()
        print("2.제목 : ", title)
        title2.append(title)
        
        area = div.find("div", "service").find("p").get_text()
        print("3.지자체이름 : ", area)
        area2.append(area)
        
        hashtag = div.find("p", "tag_type").get_text()
        print("4.해시태그 : ", hashtag)
        
        driver.find_element(By.XPATH, '//*[@id="listBody"]/ul/li[' + str(xpath_check) + ']/div[2]/div[1]/a').click()
        time.sleep(3)
        
        html_2 = driver.page_source
        soup_2 = BeautifulSoup(html_2, 'html.parser')
        contents = soup_2.find("div", id = "contents").get_text()
        print("5.본문내용 : ", contents)
        contents2.append(contents)
        
        driver.back()
        time.sleep(3)
        
        if no == collect_cnt:
            break
        
        no = no + 1
        xpath_check = xpath_check + 1
        
        if (a == 1 and xpath_check == 5):
            xpath_check = xpath_check + 1
        
    a += 1 
    b = str(a)
    xpath_check = 1
    
    if no == collect_cnt:
            break
    
    try :
        driver.find_element(By.LINK_TEXT ,'%s' % b).click()
        time.sleep(3)
    except:
        driver.find_element(By.LINK_TEXT, '다음').click()
        time.sleep(3)

print("요청하신 작업이 모두 완료되었습니다")

# Step 8. 수집된 데이터를 xls와 csv 형태로 저장하기
import os
import pandas as pd

df = pd.DataFrame()
df['번호']=no2
df['제목']=pd.Series(title2)
df['지자체']=pd.Series(area2)
df['해시태그']=pd.Series(hashtag2)
df['본문내용']=pd.Series(contents2)

n = time.localtime()
s = "%04d-%02d-%02d-%02d-%02d-%02d" % (n.tm_year, n.tm_mon, n.tm_mday, n.tm_hour, n.tm_min, n.tm_sec)
check_dic = f_dir + s + "-제주도여행"
os.makedirs(check_dic)

fx_name = check_dic + "\\" + s + "-제주도여행.xls"
fc_name = check_dic + "\\" + s + "-제주도여행.csv"

# csv 형태로 저장하기
df.to_csv(fc_name,index=False, encoding="utf-8-sig")
print("파일 저장 완료: csv 파일명 : ", fc_name)

# xls 형태로 저장하기
df.to_excel(fx_name,index=False, encoding="utf-8" , engine='openpyxl')
print("파일 저장 완료: xls 파일명 : ", fx_name)

print('요청하신 데이터 수집 작업이 정상적으로 완료되었습니다')
```
