# RHINO 사용

```python
import rhinoMorph
rn = rhinoMorph.startRhino()
text = "한글로 된 한글텍스트를 분석하는 것은 즐겁다."
```

```python
# 모든 형태소 보이기
text_analyzed = rhinoMorph.onlyMorph_list(rn, text)
print("형태소 분석 결과 : ", text_analyzed)
```

```python
# 실질형태소만, 동사의 어말어미는 제외
text_analyzed = rhinoMorph.onlyMorph_list(rn, text, pos = ['NNG', 'NNP', 'NP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'])
print("형태소 분석 결과 : ", text_analyzed)
```

```python
# 실질형태소만, 동사의 어말어미 포함
text_analyzed = rhinoMorph.onlyMorph_list(rn, text, pos = ['NNG', 'NNP', 'NP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], eomi = True)
print("형태소 분석 결과 : ", text_analyzed)
```

```python
# 전체형태소, 품사정보도 가져오기
morphs, poses = rhinoMorph.wholeResult_list(rn, text)
print("형태소 분석 결과 : ")
print("morphs : ", morphs)
print("poses : ", poses)
```

```python
# 원문의 어절 정보를 같이 가져오기
text_analyzed = rhinoMorph.wholeResult_text(rn, text)
print("형태소 분석 결과 : ", text_analyzed)
```

```python
# 명사와 명사의 연쇄는 하나의 명사로 결합하기
text_analyzed = rhinoMorph.onlyMorph_list(rn, text, pos = ['NNG', 'NNP', 'NP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], combineN = True)
print("형태소 분석 결과 : ", text_analyzed)
```

```python
# 명사와 명사의 연쇄는 하나의 명사로 결합하기
morphs, poses = rhinoMorph.wholeResult_list(rn, text, combineN = True)
print("형태소 분석 결과 : ")
print("morphs : ", morphs)
print("poses : ", poses)
```

```python
# 어근을 동사로 만들기
text_analyzed = rhinoMorph.wholeResult_list(rn, "사랑합니다.")
print("형태소 분석 결과 : ", text_analyzed)
```

```python
# 어근을 동사로 만들기
text_analyzed = rhinoMorph.wholeResult_list(rn, "사랑합니다.", xrVv = True)
print("형태소 분석 결과 : ", text_analyzed)
```

## 데이터 로딩

```python
def read_data(filename, encoding = "cp949"):
  with open(filename, "r", encoding = encoding) as f:
    data = [line.split('\t') for line in f.read().splitlines()]
    data = data[1:]
  return data

def write_data(data, filename, encoding = "cp949"):
  with open(filename, "w", encoding = encoding) as f:
    f.write(data)

data = read_data("C:\\Users\\Windows\\Desktop\\대학교\\4학년 여름방학\\Big_AI\\자연어처리\\pytest\\ratings_small.txt", encoding = 'cp949')
```

# Counter

```python
from collections import Counter

count = Counter(["여름", "과일", "봄", "딸기", "과일", "봄", "딸기", "봄"])
result = count.most_common(3)

print("count : ", count)
print("result : ", result)
print("result[0] : ", result[0])
print("result[0][0] : ", result[0][0])
print("result[0][1] : ", result[0][1])
```

```python
data_text_freq = Counter(data_text) # 적절하지 않음
print("data_text : ", data_text_freq)

mergedText = ' '.join(data_text)
print("mergedText : ", mergedText)

mergedTextList = mergedText.split(" ") # 적절함
print("mergedTextList : ", mergedTextList)
```

# Sorted

```python
# 저빈도 단어는 상관없다. 데이터 분석은 양으로 밀고간다.
sample = {"여름" : 1, "과일" : 2, "딸기" : 3}

print(sorted(sample)) # ㄱ, ㄷ, ㅇ
print(sorted(sample, reverse = True)) # ㅇ, ㄷ, ㅋ 
print(sorted(sample, key = sample.get, reverse = True)) # data 값에 따른 key 정렬
print(sorted(sample.values(), reverse = True)) # data 값에 따른 value 정렬
```

# 연습문제

```python
import rhinoMorph
rn = rhinoMorph.startRhino() # 형태소분석기 객체
```

```python
f = open("C:\\Users\\Windows\\Desktop\\대학교\\4학년 여름방학\\Big_AI\\자연어처리\\pytest\\data\\김소월시.txt", "r", encoding="utf-8")
Kim = []
for line in f.read().splitlines():
  data = line.split(" ")
  data = [d for d in data if d]
  if data:
    Kim.extend(data)
Kim = " ".join(Kim)
f.close()
```

```python
text_analyzed = rhinoMorph.onlyMorph_list(rn, Kim, pos = ['NNG', 'NNP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], eomi = True)
remove_data = ["김", "소", "월"]

text_analyzed = [data for data in text_analyzed if data not in remove_data]
```

```python
import numpy as np
from collections import Counter

count = Counter(text_analyzed)
data_keys = sorted(count, key=count.get, reverse = True)
data_values = sorted(list(count.values()), reverse = True)
sum_data = np.sum(list(count.values()))
for i in range(len(data_values)):
  data_values[i] = round(data_values[i] * 100 / sum_data, 2)

print("김소월 단어빈도 총합 : ", len(text_analyzed))
print("김소월 고빈도 단어 : ", data_keys)
print("김소월 고빈도 단어비율 : ", data_values)
```

```python
import matplotlib.pyplot as plt

print("=== 김소월 단어비율 ===")
plt.bar(range(20), data_values[:20])
plt.xticks(range(20), data_keys[:20])
plt.show()
```
