# Functional API

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Model
from keras.layers import Input, Dense
from keras.optimizers import SGD
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [1, 3, 4, 6, 8, 7, 10, 13, 14]

inputs = Input(shape = (1,))
output = Dense(1, activation = 'linear')(inputs)

model = Model(inputs, output)
model.compile(optimizer = SGD(lr = 0.01), loss = 'mse', metrics = 'acc')
model.summary()
```

```python
model.fit(x, y, epochs = 300)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Model
from keras.layers import Input, Dense, concatenate
from keras.optimizers import SGD
```

```python
x1 = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
x2 = np.array([[0, 0, 0, 0], [0, 1, 1, 0], [1, 0, 0, 1], [1, 1, 1, 1]])
y1 = np.array([0, 0, 0, 1])

x_in = Input(shape = (2,))
x_hi = Dense(4, activation = 'relu')(x_in)
x_re = Dense(2, activation = 'relu')(x_hi)
x = Model(x_in, x_re)

y_in = Input(shape = (4,))
y_hi = Dense(4, activation = 'relu')(y_in)
y_hi = Dense(2, activation = 'relu')(y_hi)
y_re = Dense(2, activation = 'relu')(y_hi)
y = Model(y_in, y_re)

z_in = concatenate([x_re, y_re])
z_hi = Dense(2, activation = 'relu')(z_in)
z_re = Dense(1, activation = 'sigmoid')(z_hi)
z = Model([x_in, y_in], z_re)

z.summary()
```

```python
z.compile(optimizer = SGD(lr = 0.01), loss = 'binary_crossentropy', metrics = ['acc'])
```

```python
z.fit([x1, x2], y1, epochs = 10)
```

```python
z.predict([x1, x2])
```

# Deep Belief Net

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, LeakyReLU
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([0, 1, 1, 0])
```

```python
model = Sequential()
model.add(Dense(2, activation = 'relu', input_shape = (2,)))
model.add(Dense(1, activation = 'sigmoid'))
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['acc'])
```

```python
model.fit(X, Y, epochs = 1000)
```

```python
X_i = model.predict(X)
```

```python
weights = model.get_weights()
```

```python
new = Sequential()
leaky_relu = LeakyReLU(alpha = 0.1)
new.add(Dense(2, activation = leaky_relu, input_shape = (2,)))
```

```python
new_i = Sequential()
leaky_relu = LeakyReLU(alpha = 0.1)
new_i.add(Dense(2, activation = leaky_relu, input_shape = (2,)))
```

```python
weights = new.get_weights()
new_i.set_weights(weights)
```

```python
new_i.compile(optimizer = 'sgd', loss = 'mse')
new_i.fit(X_i, X, epochs = 50)
```

# Weight 초기화

```python
import numpy as np
import matplotlib.pyplot as plt
from keras import optimizers
from keras.layers import Input, Dense, Dropout
from keras import initializers
from keras.models import Model
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([0, 1, 1, 0])

inputs = Input(shape = (2,))
hidden = Dense(2, activation = 'relu', kernel_initializer = initializers.GlorotNormal())(inputs)
output = Dense(1, activation = 'sigmoid', kernel_initializer = initializers.GlorotNormal())(hidden)
model = Model(inputs, output)
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['acc'])
model.summary()
```

```python
model.fit(X, Y, epochs = 500)
```

```python
model.predict(X)
```

```python
inputs = Input(shape = (2,))
hidden = Dense(2, activation = 'relu', kernel_initializer = initializers.he_normal())(inputs)
output = Dense(1, activation = 'sigmoid', kernel_initializer = initializers.he_normal())(hidden)
model = Model(inputs, output)
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['acc'])
model.summary()
```

```python
model.fit(X, Y, epochs = 500)
```

```python
model.predict(X)
```

# Feature 정규화

```python
import numpy as np
import matplotlib.pyplot as plt
from keras import optimizers
from keras.layers import Input, Dense, Dropout, BatchNormalization
from keras.models import Model
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([0, 1, 1, 0])

inputs = Input(shape = (2,))
bn = BatchNormalization()(inputs)
hidden = Dense(2, activation = 'relu')(bn)
output = Dense(1, activation = 'sigmoid')(hidden)
model = Model(inputs, output)
model.compile(optimizer = 'sgd', loss = 'binary_crossentropy', metrics = ['acc'])
```

```python
model.fit(X, Y, epochs = 50)
```

```python
model.predict(X)
```
